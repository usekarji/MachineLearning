```python
#Data Clensing 

def fillNullWithMedian(df,column):
    df[column] = df[column].fillna(df[column].dropna().median())

def clean_data(data):
    fillNullWithMedian(data,'Fare')
    fillNullWithMedian(data,'Age')
    data.loc[data['Sex']=='male',"Sex"] = 1
    data.loc[data['Sex']=='female',"Sex"] = 0
    
    data["Embarked"] = data["Embarked"].fillna("S")
    data.loc[data["Embarked"]=="S","Embarked"]= 0
    data.loc[data["Embarked"]=="C","Embarked"]= 1
    data.loc[data["Embarked"]=="Q","Embarked"]= 2 
    return data

def generateVIF(X_tr):
    df = pd.DataFrame()
    df['Features'] = X_tr.columns
    df['VIF'] = [variance_inflation_factor(X_tr.values, i) for i in range(X_tr.shape[1])]
    df = df.sort_values(by = "VIF", ascending = False)
    return(df)
    
```

```python
train["hyp"] = 0
train.loc[train.Sex=="female","hyp"]=1

train["result"] = 0
train.loc[train.Survived == train["hyp"], "result"]=1 
train["result"].value_counts(normalize=True)
train = train.drop(columns=['PassengerId','Name','Parch','Ticket','Cabin'])
train1 = clean_data(train)
train1.head()
```

```python
y_train = train1.pop('result')
X_train = train1

#Creating linear regression 

lm = LinearRegression()
lm.fit(X_train,y_train)
# Selecting the top 5 features using RF elemination
rfe = RFE(lm, 5)
rfe = rfe.fit(X_train, y_train)
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
X_train.columns[rfe.support_]
X_train_RFE = X_train[X_train.columns[rfe.support_]]
X_train_RFE.head()
```

```python
X_train_RFE.corr()
```

```python
#Building the model using statsmodel
X = sm.add_constant(X_train_RFE)
lm = sm.OLS(y_train,X).fit()
lm.summary()
```

```python
generateVIF(X)
```

```python
X1 = sm.add_constant(X)
lm = sm.OLS(y_train,X1).fit()
lm.summary()
```

```python
generateVIF(X1)

LinearModel = sm.OLS(y_train,X1).fit()
y_train_pred = LinearModel.predict(X1)

#Plotting the error terms 
fig = plt.figure()
fig.suptitle("Error Terms Distribution")
sns.distplot((y_train - y_train_pred), bins = 10)
```
